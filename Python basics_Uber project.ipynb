## Q19. Plot a bar graph of Purpose vs Miles(Distance). What can you infer from the plot(2 +2 points)
#### Note: Use the original dataframe without dropping the 'NA' values.
#### Hint:You have to plot total/sum miles per purpose

```python
pmdf = pd.DataFrame(ubdf.groupby('PURPOSE*')['MILES*'].sum()).sort_values(by='PURPOSE*',ascending=False).head()
pmdf = pmdf.reset_index()
plt.figure(figsize=(15,7))
sns.barplot(pmdf['PURPOSE*'],pmdf['MILES*']);
```

## Q22. What percentage of Miles were clocked under Business Category and what percentage of Miles were clocked under Personal Category ? (3 points)

### Note:Use the original dataframe without dropping the 'NA' values. 


```python
catmiles = pd.DataFrame(ubdf.groupby('CATEGORY*')['MILES*'].sum()).sort_values(by='CATEGORY*',ascending=False).head()
catmiles
```

```python
a = catmiles['MILES*'].sum()
a
```

```python
for i in catmiles['MILES*'].unique():
    print("Percentage is",i/a*100 )
    
```

Personal miles clocked is 5.88% and Business is 94.11%
